```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.style.use('ggplot')
```

```python
df['size'] = df['size'].str.split(' ').str[0]
df['size'].fillna(df['size'].mode()[0],inplace=True)
df
```

```python
df['balcony'].fillna(df['balcony'].mode()[0],inplace=True)
```

```python
df['bath'].fillna(df['bath'].mode()[0],inplace=True)
```

```python
df['bath'] = df['bath'].astype(int)
df['size'] = df['size'].astype(int)
df['balcony'] = df['balcony'].astype(int)
df['total_sqft'] = df['total_sqft'].str.replace(r'[^\d\.]+', '', regex=True)
df['total_sqft'] = pd.to_numeric(df['total_sqft'],errors='coerce')
df.dropna(inplace=True)
df['total_sqft'] = df['total_sqft'].astype(int)
df
```

```python
df = df[['location','total_sqft','bath','price']]
df
```

```python
X = df.drop(columns=['price'])
y = df.price
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
df
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output=),['location']),remainder='passthrough')
column_trans
```

```python
scaler = StandardScaler(with_mean=False)
scaler
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,scaler,lr)
pipe
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
y_pred
```

```python
r2_score(y_test,y_pred)
```
